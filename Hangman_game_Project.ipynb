## Description
> ##### The Hangman program randomly selects a secret word from a list of secret words.
> ##### The Game: Here, a random word (a fruit name) is picked up from our collection and the player gets limited chances to win the game.
> ##### When a letter in that word is guessed correctly, that letter position in the word is made visible. In this way,all letters of the word are to be guessed before all the chances are over.
> ##### For convenience, we have given length of word + 2 chances. For example, word to be guessed is mango, then user gets 5 + 2 = 7 chances, as mango is a five letter word.

```python
#first of all i read a many codes for this game cuz i wasnt able to understand it as a code in the beginning
#but this is my work and i hope u enjoy it 
import random
import time

#Welcome function help to welcoming the players to the game by asking for name and introduce them to the game
def welcome():
    print("Welcome to Hangman Game\n")
    name = input("Enter your name: ").strip()
    print("Hello " + name.capitalize() + "! Best of Luck!")
    time.sleep(1)
    print("The game is about to start!\nLet's play Hangman!")
    time.sleep(2)

    

#Main function is the core section of the game where the guessed word in a list and ordering the counting prosses
def main():
    global count
    global display
    global word
    global already_guessed
    global length
    global play_game
    
    words_to_guess = ["november ","bread","image","movie","epsilon","baby","heart","toy","avatar","laptop","football"]
    word = random.choice(words_to_guess)
    length = len(word)
    count = 0
    display = '_'* length
    already_guessed = []
    play_game = ""
    print("-"*50)
print("-"*50)
   

#Game_loop function is the function to keep playing or execute the game whenever you win or lose
def game_loop():
    global play_game
    
    play_game = input("Do You want to play again? y = yes, n = no \n")
    while play_game not in ["y", "n","Y","N","yes","no","Yes","No"]:
        play_game = input("Do You want to play again? y = yes, n = no \n")
    if play_game == "y" or play_game == "yes" or play_game == "Yes":
        main()
    elif play_game == "n" or play_game == "no" or play_game == "No":
        print("Thanks For Playing! We expect you back again!")
        exit()
        
        
#Hangman function is the stage of the game there you can see the length of your word and if you get the right letter or not
#Also it happen here if you lose you will be hang or if you win the game will congrate you for winning
#And finaly ask you if you want to keep playing or exit from the game 
def hangman():
    global count
    global display
    global word
    global already_guessed
    global play_game
    global length
    
    #making the limit of chances word +2
    limit = length +2
    
    print("This is the Hangman game Word, It contains " , length , " letters" )
    guess = input("Entre your guess letter : " + display+'\n')
    guess = guess.strip()
    
    if len(guess.strip()) == 0 or len(guess.strip()) >= 2 or guess <= "9":
        print("Invalid Input, Try another letter\n")
        hangman()


    elif guess in word:
        already_guessed.extend([guess])
        index = word.find(guess)
        word = word[:index] + "_" + word[index + 1:]
        display = display[:index] + guess + display[index + 1:]
        print(display + "\n")

    elif guess in already_guessed:
        print("Try another letter.\n")
        
    elif word == '_' * length:
        print("Congrats! You have guessed the word correctly!, It is " + display.capitalize() +" *-*")
        game_loop()

    

    else:
        count += 1
      

        if count == limit:
            time.sleep(1)
            print("   _____ \n"
                  "  |     | \n"
                  "  |     |\n"
                  "  |     O \n"
                  "  |    /|\ \n"
                  "  |    / \ \n"
                  "__|__\n")

            print("Wrong guess. You are hanged!!!\n")
            print("The word was:",already_guessed,word)
            game_loop()
        
        #note: i was trying hard to make the rest of this codelines in a shape of loop but i couldnt do it
        #so i hope in a comments or in session to show me a better way to do it
        
        elif count == 1:
            time.sleep(1)
            print("Wrong guess. " + str(limit - count) + " guesses remaining\n")
            
            
        elif count == 2:
            time.sleep(1)

            print("Wrong guess. " + str(limit - count) + " guesses remaining\n")    

        elif count == 3:
            time.sleep(1)

            print("Wrong guess. " + str(limit - count) + " guesses remaining\n")

        elif count == 4:
            time.sleep(1)

            print("Wrong guess. " + str(limit - count) + " guesses remaining\n")

        elif count == 5:
            time.sleep(1)

            print("Wrong guess. " + str(limit - count) + " guesses remaining\n")
            
        elif count == 6:
            time.sleep(1)

            print("Wrong guess. " + str(limit - count) + " guesses remaining\n")
            
        elif count == 7:
            time.sleep(1)

            print("Wrong guess. " + str(limit - count) + " guesses remaining\n")
            
        elif count == 8:
            time.sleep(1)

            print("Wrong guess. " + str(limit - count) + " guesses remaining\n")
        
        elif count == 9:
            time.sleep(1)

            print("Wrong guess. " + str(limit - count) + " guesses remaining\n")

        elif count == 10:
            time.sleep(1)

            print("Wrong guess. " + str(limit - count) + " guesses remaining\n")       
  
    if count != limit:
        hangman()        

        
                
welcome()

main()

hangman()
```
